```python
import numpy as np
import pickle

with open('timing_results(9).pkl', 'rb') as f:
    old_data = pickle.load(f)
old_data
```

```python
with open('timing.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
from matplotlib import pyplot as plt
```

```python
del data['Multiplication']
data['Bertini'] = old_data['bertini']
data['TVB'] = old_data['TVB power']
degrees = data['Degrees']
degrees
```

```python
data
```

```python
methods = 'Bertini', 'TVB', 'Subdivision', 'Division'
```

```python
#set up figure, set size
plt.figure(dpi=300,figsize=(3,5))
fig,ax = plt.subplots(1)
fig.set_size_inches(5,4)

plt.subplot(121)
for method in methods:
    plt.plot(degrees, data[method], label=method)
    
#titles, axes
plt.title('Linear Scale')
plt.xlabel('degree')
plt.ylabel('time (seconds)')

plt.legend()

plt.subplot(122)
for method in methods:
    plt.plot(degrees, data[method])
    
#titles, axes
plt.title('Log Scale')
plt.xlabel('degree')
plt.yscale('log')

#save
title = 'timing_both_slides'
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig(title + '_1200.png',dpi=1200, bbox_inches='tight')
plt.savefig(title + '_300.png',dpi=300, bbox_inches='tight')
```

```python
#set up figure, set size
plt.figure(dpi=1200)
fig,ax = plt.subplots(1)
fig.set_size_inches(5,6)

plt.suptitle('Timing Comparisons')

plt.subplot(121)
for method in methods:
    plt.plot(degrees, data[method], label=method)
    
#titles, axes
plt.title('Linear Scale')
plt.xlabel('degree')
plt.ylabel('time (seconds)')

plt.legend()

plt.subplot(122)
for method in methods:
    plt.plot(degrees, data[method])
    
#titles, axes
plt.title('Log Scale')
plt.xlabel('degree')
plt.yscale('log')

#save
title = 'timing_both_poster'
fig.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig(title + '_1200.png',dpi=1200, bbox_inches='tight')
plt.savefig(title + '_300.png',dpi=300, bbox_inches='tight')
```
